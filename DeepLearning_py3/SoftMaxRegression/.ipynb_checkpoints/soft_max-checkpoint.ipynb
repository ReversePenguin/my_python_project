Softmax回归算法，又称为多项式或多类别的Logistic回归算法:<br/>
z = w*x + b


激活函数:<br/>
   softmax函数，将得分值转化为概率值形式


损失函数:<br/>
   交叉熵损失函数





```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

数据集

```python
np.random.seed(123)
#中心的数量决定了y的值的范围：[0，centers-1]
X,Y = make_blobs(n_samples = 5000,n_features=5, centers=4)    
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c = Y)
plt.title("Dataset")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

```python
Y = Y[:, np.newaxis]  # 转化Y的维度，变为（n_samples，1）

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print('Shape X_train: {}'.format(X_train.shape))
print('Shape y_train: ', Y_train.shape)
print('Shape X_test: ', X_test.shape)
print('Shape y_test: ', Y_test.shape)
```

SOftmax回归类

```python
class SoftmaxRegressor:
    
    def __init__(self):
        pass
    
    def train(self,X,Y,n_classes,learning_rate = 0.01, n_iters = 1000):
        
        self.m, n = X.shape
        self.n_classes = n_classes
        self.w = np.random.rand(self.n_classes, n) * 0.01
        self.b = np.zeros((self.n_classes,1))
        all_losses = []
        
        for i in range(n_iters):            
            soft_z = self.softmax(X)     #(n_classes,m)
            y_predict = np.argmax(soft_z,axis=0)[:,np.newaxis]     #(m,1)
            
            y_one_hot = self.one_hot(y_predict)     #(n_classes,m)
            loss = self.compute_loss(y_one_hot,soft_z)
            all_losses.append(loss)
            
            #计算w和b的偏导
            dw = (1/self.m) * np.dot((soft_z - y_one_hot),X)    #(n_classes,n)
            db = (1/self.m) * np.sum(soft_z - y_one_hot,axis=1,keepdims=True)     #(n_classes,1)
            #print(db.shape)
            
            #梯度下降法，更新w和b
            self.w = self.w - learning_rate * dw
            self.b = self.b - learning_rate * db
            
            if i % 100 == 0:
                print(f'Iteration number:{i},loss:{np.round(loss,4)}')
                
        return self.w,self.b,all_losses
            
            
    
    def compute_loss(self,Y,soft_z):
        loss = -(1/self.m) * np.sum(Y * np.log(soft_z))
        return loss
        
    
    def softmax(self,X):
        #计算softmax的值
        #X:(m,n)
        z = np.dot(self.w,X.T) + self.b     #(n_classes,m)
        exp = np.exp(z)      #(n_classes,m)
        sum_exp = np.sum(exp,axis=0, keepdims = True)    #(1,m)
        
        return exp / sum_exp
    
    def one_hot(self,y):
        #把每一个样本中概率最大的那个y值设置为1，其他为0
        one_hot = np.zeros((self.n_classes,self.m))
        one_hot[y.T,np.arange(self.m)] = 1
        return one_hot
    
    
    def predict(self,X):
        #X:(m,n)
        soft_z = self.softmax(X) 
        #最大值的下标就是y的预测值
        return np.argmax(soft_z,axis=0)[:,np.newaxis]
        
    
    
    
```

初始化并训练模型

```python
regressor = SoftmaxRegressor()
n_classes = 4
learning_rate = 0.01
n_iters = 800
w_trained, b_trained, loss = regressor.train(X_train,Y_train,n_classes,learning_rate, n_iters)


fig = plt.figure(figsize=(8,6))
plt.plot(np.arange(n_iters),loss)
plt.title("Development of loss during training")
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()



```

```python
m_test = X_test.shape[0]
y_predict = regressor.predict(X_test)
print(y_predict[:15])

print(f'Classification accuracy on test set:{(np.sum(y_predict == Y_test)/m_test)*100}%')














```
